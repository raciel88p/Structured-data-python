Ahora vamos a crear una nueva columna donde vamos a darle un valor categorico

```python
def faang_classifer(company):
  faang = ['Facebook', 'Amazon', 'Apple', 'Netflix', 'Google']
  if company in faang:
    return 1
  else:
    return 0
```

Ingresando la nueva columna de la empresa, se puede usar como filtro

```python
df['Is_Faang'] = df['company'].apply(lambda x: faang_classifer(x))
```

Graficos con estimacion de desidad kernel

```python
plt.figure(figsize=(25, 10))

#manipulamos el primer grafico de la izquierda
plt.subplot(1, 2, 1)
sns.kdeplot(data = df, x = 'basesalary', cumulative=True)
plt.title('Densidad de kernel basado en basesalary')

#manipulamos el segundo grafico de la izquierda
plt.subplot(1, 2, 2)
sns.kdeplot(data = df, x = 'Is_Faang', cumulative=True)
plt.title('Densidad de kernel basado en Is_Faang')

#location is not posible becasue is not a number

plt.show()
```

Tarea del proyecto 3 Crear una nube de palabras de mi preferencia (La ubicacion de los usuarios)

```python
from wordcloud import WordCloud
```

```python
#creamos un variable llamada tag_words 
tag_words = ''
stop_words = {'nan', 'title'}

for i in df.location:
  i = str(i)
  separate = i.split()
  #print(separate) para saber que esta leyendo toda la tabla en location y si lo esta haciendo
  #lo siguiente es para hacer toda las palabras en minuscula
  #for j in range(len(separate)):
    #separate[j] = separate[j].lower()
  
  tag_words += ' '.join(separate)+ ' '
  print(separate)



#aca tengo la duda de porque no se me visualiza los demas lugares
```

```python
tag_words
```

Creamos la nube de palabras

```python
wordcloud_1 = WordCloud(width = 800, height= 800, background_color= 'white', stopwords= stop_words,
                        min_font_size= 10).generate(tag_words)

plt.figure(figsize = (10,10))
plt.imshow(wordcloud_1)
plt.axis('off')

plt.show()
```
